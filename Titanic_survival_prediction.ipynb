```python
#loading dataset
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
#exploring the data
df.head()
```

```python
df.describe()
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#Data Preprocessing
#handling missing values
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.drop('Cabin',axis=1,inplace=True)
```

```python
#Converting categorical features to numeric
df['Sex']=df['Sex'].map({'male':0,'female':1})
df['Embarked']=df['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
#Conditional dropping unnecessary columns
print(df.columns)
columns_to_drop=['Name','Ticket']
existing_columns_to_drop=[col for col in columns_to_drop if col in df.columns]
if existing_columns_to_drop:
  df.drop(existing_columns_to_drop,axis=1,inplace=True)
else:
  print(f"Columns{columns_to_drop}not found in the DataFrame")
```

```python
#verifying changes
print(df.head())
```

```python
#Defining features and target variables
print("Column Names in the dataset:",df.columns)

target_column='Survived'
if target_column in df.columns:
 x=df.drop(target_column,axis=1)
 y=df[target_column]
 print("Feature DataFrame(x) Head:\n",x.head())
 print("Target variable(y) Head:\n", y.head())
else:
  raise KeyError(f"The target column '{target_column}' for survival status is not found in the dataset.")

```

```python
#splitting data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Building the model
#importing classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#initializing and training the model
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
```

```python
#make predictions
y_pred=model.predict(x_test)
```

```python
#evaluating the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
#Finding Accuracy
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy:{accuracy}')
#printing classification report
print('Classification Report:')
print(classification_report(y_test,y_pred))
#printing Confusion Matrix
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```
